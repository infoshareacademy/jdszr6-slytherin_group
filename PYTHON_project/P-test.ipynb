Chi-square test
source: https://thinkingneuron.com/how-to-measure-the-correlation-between-two-categorical-variables-in-python/

how gender impacts credit history

```python
crosstabResult=pd.crosstab(index=df["CODE_GENDER"],columns=df["STATUS"])
crosstabResult
```

```python
from scipy.stats import chi2_contingency
```

```python
ChiSqResult = chi2_contingency(crosstabResult)
```

```python
ChiSqResult[1]
```

how car ownership impacts credit history

```python
crosstabResultFOC=pd.crosstab(index=df["FLAG_OWN_CAR"],columns=df["STATUS"])
crosstabResultFOC
ChiSqResultFOC = chi2_contingency(crosstabResultFOC)
ChiSqResultFOC[1]
```

```python
crosstabResultFOC
```

how realty ownership impacts credit history

```python
crosstabResultFOR=pd.crosstab(index=df["FLAG_OWN_REALTY"],columns=df["STATUS"])
crosstabResultFOR
ChiSqResultFOR = chi2_contingency(crosstabResultFOC)
ChiSqResultFOR[1]
```
